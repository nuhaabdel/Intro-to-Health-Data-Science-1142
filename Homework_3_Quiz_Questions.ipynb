**What is the primary goal of data anonymization?**

To conceal or remove information that might be used to identfy someone in order to protect their privacy.

**What is the Havasupai Tribe case about?**

Researchers broke trust and informed consent by using the Havasupai tribe's blood samples for research for which they had not given consent to.  

**What was the largest health data breach in history?**

An estimated 78.8 million people's personal information was compromised in the Anthem data breach in 2015.

**What is the pulse oximeter controversy about?**

Because pulse oximeters were mainly developed using data from people with lighter skin tones, they may be less reliable for Black patients.

**What is the role of an Institutional Review Board (IRB) in research?**

To examine and keep an eye on human subjects research to ensure that it is morally sound and protects participants.

**What is the difference between spreadsheet software and database software for health data management?**

Spreadsheets are straightforward tables for modest datasets. Databases are large complicated datasets, which also have safe storage and improved organization.

**What is a conflict of interest in public health research?**

When a researcher's financial, professional, or personal interests potentially affect the findings of their study.

**What is informed consent in the context of health data collection?**

When individuals willingly consent to the use of their data after being properly informed about it.

**What is data masking in the context of de-identification of health data?**

Hiding identifying information to prevent people from being recognized.

**What is the principle of k-anonymity in the context of de-identification of health data?**

Each individual in the dataset cannot be identified from at least k other individuals due to the grouping of the data.

**What is the main concern about conflicts of interest in public health research?**

They have the potential to influence research findings and reduce trust in them.

**What is the first step in managing conflicts of interest in public health research?**

Transparency and disclosure of potential conflicts.

**What is the role of independent oversight in managing conflicts of interest in public health research?**

To make sure conflicts dont impact the findings, a separate team evaluares and keeps an eye on the research.

**What is one strategy for mitigating conflicts of interest in public health research?**

Public disclosure of the conflict, or avoiding it.

**What is the potential impact of not properly managing conflicts of interest in public health research?**

It can lead to biased research, loss of credibility, and loss of public trust.
